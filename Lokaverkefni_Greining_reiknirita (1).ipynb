### Búa til netið

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, node1, bus_node, node2, departure_time, arrival_time):
        if node1 not in self.graph:
            self.graph[node1] = {}
        if bus_node not in self.graph:
            self.graph[bus_node] = {}
        if node2 not in self.graph:
            self.graph[node2] = {}
        self.graph[node1][bus_node] = departure_time
        self.graph[bus_node][node2] = arrival_time





g = Graph()

for index, row in dfnytt.iterrows():
    node1 = str(row['departure_stop_id'])
    bus_node = row['leg_id']
    node2 = str(row['arrival_stop_id'])
    departure_time = row['departure_time']
    arrival_time = row['arrival_time']

    g.add_edge(node1, bus_node, node2, departure_time, arrival_time)
```

### 2.3.2 Leit og 2.3.3 Framsetning

```python
import heapq

# breyta tíma í mínútur eftir miðnætti til að hægt sé að bera saman tíma
def time_to_minutes(t):
    if pd.isna(t):
        return float('inf')  # Ef NaN gildi
    try:
        h, m, s = map(int, t.split(':'))
        return h * 60 + m + s / 60

    except ValueError:
        return float(t)  # Ef gildið er nú þegar mínútur þá skilum við mínútunum (notað í seinni liðum)


def minutes_to_time(minutes):
    if minutes == float('inf'):
        return None  # None ef inntakið er inf
    h = int(minutes // 60)
    m = int(minutes % 60)
    s = int((minutes * 60) % 60)
    return f"{h:02d}:{m:02d}:{s:02d}"




# finna stop_id fyrir þessi stop_name
def find_shortest_path(g, source, target, departure_time_str):
    shortest_path = []
    shortest_length = float('inf')

    source_ids_raw = stop_name_to_stop_ids[source]
    target_ids_raw = stop_name_to_stop_ids[target]

    # sum stop_id eru ekki í netinu
    source_ids = [stop_id for stop_id in source_ids_raw if str(stop_id) in g]
    target_ids = [stop_id for stop_id in target_ids_raw if str(stop_id) in g]

    # Nota Dijkstra til að finna stystu leið
    def Dijkstra1(g, source_id, target_id, departure_time_str):
        departure_time = time_to_minutes(departure_time_str)  # Breyta tíma úr streng í mínútur frá miðnætti

        dist = {node: float('inf') for node in g}
        dist[source_id] = departure_time

        pq = []
        heapq.heappush(pq, (departure_time, source_id))

        pred = {}

        final_time = float('inf')
        found = False

        while pq:
            t, u = heapq.heappop(pq)

            # þarf að gera þetta þar sem að það er ekki hægt að gera decreaseKey
            if t != dist[u]:
              continue

            if u == target_id:
                final_time = t
                found = True
                break

            for v in g[u]:
              wt = time_to_minutes(g[u][v])

              # leggurinn er ólöglegur
              if wt < dist[u]:
                continue

              if wt < dist[v]:
                dist[v] = wt
                pred[v] = u

                heapq.heappush(pq, (dist[v], v))


        leid = []

        if found:
          # finna leiðina
          while target_id is not None:
              leid.append(target_id)
              target_id = pred.get(target_id)

          leid.reverse()

        return leid, final_time # inf ef við komumst ekki


    for source_id in source_ids:
        for target_id in target_ids:
            if source_id != target_id:
                path, final_time = Dijkstra1(g, str(source_id), str(target_id), departure_time_str)

                if path != [] and final_time < shortest_length:
                    shortest_path = path
                    shortest_length = final_time

    return shortest_path, shortest_length




source_stop_name = 'Meistaravellir'
target_stop_name = 'FB'
initial_departure_time = "11:30:00"

leid, result_time = find_shortest_path(g.graph, source_stop_name, target_stop_name, initial_departure_time)

def prenta_leid(leid, result_time):
  first_trip = leid[1].split('_')[0]
  straetoNr = leidir[int(first_trip)]
  trip_headsign = trips[trips['trip_id'] == int(first_trip)]['trip_headsign'].item()
  stoppNafn = stop_id_to_stop_name[int(leid[0])]
  time = g.graph[leid[0]][leid[1]]

  print(f"Á stoppistöðinni {stoppNafn}: Taka strætó nr. {straetoNr} -> {trip_headsign} sem fer kl. {time}. Hann fer þessa leið: \n")

  for i, element in enumerate(leid):
      if  i == 0:
        continue

      if i % 2 == 0:
        print(f"{stop_id_to_stop_name[int(element)]} ({g.graph[leid[i-1]][leid[i]]})\n")

      if i != 0 and i != len(leid)-1 and not ('_' in leid[i]):
        if (leid[i-1].split('_')[0] != leid[i+1].split('_')[0]):

          next_trip = leid[i+1].split('_')[0]
          straetoNr = leidir[int(next_trip)]
          trip_headsign = trips[trips['trip_id'] == int(next_trip)]['trip_headsign'].item()
          stoppNafn = stop_id_to_stop_name[int(element)]
          time = g.graph[leid[i]][leid[i+1]]

          print(f"Á stoppistöðinni {stoppNafn}: Skipta yfir í strætó nr. {straetoNr} -> {trip_headsign} sem fer kl. {time}. Hann fer þessa leið: \n")


  print(f"Komutími er {minutes_to_time(result_time)} \n")


if result_time != float('inf'):
    prenta_leid(leid, result_time)
else:
    print("Engin leið fannst")
```

### 2.3.4 Labb

Til að gera þennan hluta þá þurfum við að byrja á að finna stoppustöðvarnar sem eru í göngufjarlægð og bæta þá legg á milli þeirra sem tákna tíma sem tekur að ganga á milli í mínútum.

```python
class Graph_labb:
    def __init__(self):
        self.graph = {}

    def add_edge(self, node1, bus_node, node2, departure_time, arrival_time):
        if node1 not in self.graph:
            self.graph[node1] = {}
        if bus_node not in self.graph:
            self.graph[bus_node] = {}
        if node2 not in self.graph:
            self.graph[node2] = {}
        self.graph[node1][bus_node] = (departure_time, 'bus')
        self.graph[bus_node][node2] = (arrival_time, 'bus')

    def add_walking_edge(self, node1, node2, walking_time):
      if node1 in self.graph and node2 in self.graph:
        self.graph[node1][node2] = (walking_time, 'walk')


g_labb = Graph_labb()

for index, row in dfnytt.iterrows():
    node1 = str(row['departure_stop_id'])
    bus_node = row['leg_id']
    node2 = str(row['arrival_stop_id'])
    departure_time = row['departure_time']
    arrival_time = row['arrival_time']

    g_labb.add_edge(node1, bus_node, node2, departure_time, arrival_time)
```

```python
import math

# Þetta er kannski smá overkill
def great_circle(lat1, lon1, lat2, lon2):

    # Breyta í radíana
    lon1, lat1, lon2, lat2 = map(math.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    # formúla Haversine til að finna fjrlægðina á milli punktanna
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))

    r = 6371  # Radíus jarðar í km
    return c * r



# Bætir við göngu legg ef þeir eru innan við ákveðins vegalengdar
def add_walking_edges(graph, stops, max_walking_distance):

    for index1, stop1 in stops.iterrows():

        for index2, stop2 in stops.iterrows():

            if stop1['stop_id'] != stop2['stop_id']:
                distance = great_circle(stop1['stop_lat'], stop1['stop_lon'], stop2['stop_lat'], stop2['stop_lon'])

                if distance <= max_walking_distance:
                    # Göngutími í mínútum (námundað upp)
                    walking_time = math.ceil((distance / 5.0) * 60)  # Grf að gönguhraði sé 5km/klst

                    graph.add_walking_edge(str(stop1['stop_id']), str(stop2['stop_id']), walking_time)


# max 0.5 km labb milli stoppistöðva
add_walking_edges(g_labb, stops, 0.5)
```

```python
# breyta aðeins fallinu sem leitar að stystu ferð

def find_shortest_path_labb(g, source, target, departure_time_str):
    shortest_path = []
    shortest_length = float('inf')

    source_ids_raw = stop_name_to_stop_ids[source]
    target_ids_raw = stop_name_to_stop_ids[target]

    # sum stop_id eru ekki í netinu
    source_ids = [stop_id for stop_id in source_ids_raw if str(stop_id) in g]
    target_ids = [stop_id for stop_id in target_ids_raw if str(stop_id) in g]

    def Dijkstra2(g, source_id, target_id, departure_time_str):
        departure_time = time_to_minutes(departure_time_str)

        dist = {node: float('inf') for node in g}
        dist[source_id] = departure_time

        pq = []
        heapq.heappush(pq, (departure_time, source_id))

        pred = {}

        final_time = float('inf')
        found = False

        while pq:
            t, u = heapq.heappop(pq)

            if t != dist[u]:
                continue

            if u == target_id:
                final_time = t
                found = True
                break

            for v, (wt, mode) in g[u].items():
                if mode == 'walk':
                    next_time = dist[u] + float(wt)
                else:
                    wt_minutes = time_to_minutes(wt)

                    # leggurinn er ólöglegur
                    if wt_minutes < dist[u]:
                      continue

                    next_time = wt_minutes

                if next_time < dist[v]:
                    dist[v] = next_time
                    pred[v] = u

                    heapq.heappush(pq, (dist[v], v))

        if found:
            leid = []
            step = target_id
            while step is not None:
                leid.insert(0, step)
                step = pred.get(step)

            return leid, final_time

        return [], float('inf')


    for source_id in source_ids:
        for target_id in target_ids:
            if source_id != target_id:
                path, final_time = Dijkstra2(g, str(source_id), str(target_id), departure_time_str)

                if path != [] and final_time < shortest_length:
                    shortest_path = path
                    shortest_length = final_time

    return shortest_path, shortest_length




source_stop_name = 'Meistaravellir'
target_stop_name = 'FB'
initial_departure_time = "11:30:00"


path, result_time = find_shortest_path_labb(g_labb.graph, source_stop_name, target_stop_name, initial_departure_time)




# breyta aðeins fallinu sem býr til framsetninguna

def prenta_leid_labb(leid, g, final_time):
    if not leid:
        print("Engin leið fannst")
        return

    i = 0

    while i < len(leid)-1:
      node = leid[i]
      next_node = leid[i+1]

      # labb
      if '_' not in node and '_' not in next_node:
        walking_time = g[leid[i]][leid[i+1]][0]
        stop_name = stop_id_to_stop_name[int(node)]
        next_stop_name = stop_id_to_stop_name[int(next_node)]

        print(f"Á stoppistöðinni {stop_name}: Ganga til stoppistöðvarinnar {next_stop_name}, tekur {walking_time} mín. \n")

        i = i+1
      else:
        next_trip = next_node.split('_')[0]
        straetoNr = leidir[int(next_trip)]
        trip_headsign = trips[trips['trip_id'] == int(next_trip)]['trip_headsign'].item()
        stoppNafn = stop_id_to_stop_name[int(node)]
        time = g[leid[i]][leid[i+1]][0]

        print(f"Á stoppistöðinni {stoppNafn}: Taka strætó nr. {straetoNr} -> {trip_headsign} sem fer kl. {time}. Hann fer þessa leið: \n")

        i = i+2

        while leid[i-1].split('_')[0] == leid[i+1].split('_')[0]:
          print(f"{stop_id_to_stop_name[int(leid[i])]} ({g[leid[i-1]][leid[i]][0]})\n")
          i = i+2

        print(f"{stop_id_to_stop_name[int(leid[i])]} ({g[leid[i-1]][leid[i]][0]})\n")


    print(f"Komutími er kl. {minutes_to_time(final_time)}")



if result_time != float('inf'):
    prenta_leid_labb(path, g_labb.graph, result_time)
else:
    print("Engin leið fannst")
```

### 2.3.6 Bestun og gagnagrindur

Í staðinn fyrir að keyra Dijkstra fyrir alla hnútana að þá skiluðum við úr fallinu um leið og við fengum endahnútinn úr forgangsbiðröðinni (enda engin ástæða þá að halda áfram að skoða fleiri hnúta). Við notuðum uppflettitöflur þegar það þurfti t.d. að fletta upp stop_id fyrir stop_name, í staðinn fyrir að sækja úr pandas dataframe.

Ein hugmynd til að bæta afköst reikniritisins var að breyta netinu þ.a. við myndum hafa einn hnút fyrir hvert nafn á stoppistöð þar sem að stoppistöðvar með sama nafn eru mjög nálægt hvor annarri. Þetta myndi fækka hnútum en hins vegar er meirihluti hnútanna tilkominn vegna strætóferðanna. Einnig tókum við eftir því að í einni leitarniðurstöðu á heimasíðu strætó þurfti að skipta um strætó með því að ganga frá stoppistöð til annarrar stoppistöðvar sem hét sama nafni. Þannig að ef við hefðum gert þetta hefðum við glatað þessum upplýsingum.

### 2.3.7 Færri skiptingar

Við gerum þetta með því að halda um fjölda ólíkra trip_id fyrir hverja leið að hverju sinni í Dijkstra. Ef það kemur síðan upp jafntefli í Dijkstra fyrir tvær mögulegar leiðir (next_time == dist[v]) að þá berum við sama fjölda ólíkra trip_id og uppfærum leiðina ef fjöldi ólíkra trip_id fyrir nýju leiðina er minni

```python
# breyta aðeins fallinu sem finnur stystu leið

def find_shortest_path_labb_faerri_skiptingar(g, source, target, departure_time_str):
    shortest_path = []
    shortest_length = float('inf')

    source_ids_raw = stop_name_to_stop_ids[source]
    target_ids_raw = stop_name_to_stop_ids[target]

    # sum stop_id eru ekki í netinu
    source_ids = [stop_id for stop_id in source_ids_raw if str(stop_id) in g]
    target_ids = [stop_id for stop_id in target_ids_raw if str(stop_id) in g]


    def Dijkstra3(g, source_id, target_id, departure_time_str):
        departure_time = time_to_minutes(departure_time_str)  # Convert departure time to minutes from midnight

        dist = {node: float('inf') for node in g}
        dist[source_id] = departure_time

        pq = []
        heapq.heappush(pq, (departure_time, source_id))

        pred = {}

        node_to_trip_ids = {}
        node_to_trip_ids.setdefault(source_id, set())

        final_time = float('inf')
        found = False

        while pq:
            t, u = heapq.heappop(pq)

            if t != dist[u]:
                continue

            if u == target_id:
                final_time = t
                found = True
                break

            for v, (wt, mode) in g[u].items():
                if mode == 'walk':
                    next_time = dist[u] + float(wt)
                else:
                    wt_minutes = time_to_minutes(wt)

                    # leggurinn er ólöglegur
                    if wt_minutes < dist[u]:
                      continue

                    next_time = wt_minutes

                if next_time < dist[v]:
                    dist[v] = next_time
                    pred[v] = u

                    heapq.heappush(pq, (dist[v], v))

                    # hnúturinn v fær öll þau ólíku trip_id sem u hafði
                    node_to_trip_ids[v] = node_to_trip_ids[u].copy()

                    # ef hnútur táknar ferð, bæta við því trip_id
                    if '_' in v:
                      trip_id = v.split('_')[0]
                      node_to_trip_ids.setdefault(v, set()).add(trip_id)

                # ef jafntefli fyrir tíma, þá berum við saman fjölda ólíkra trip_id
                elif next_time == dist[v]:
                  total_trips_v = len(node_to_trip_ids[v])
                  total_trips_u = len(node_to_trip_ids[u])


                  if total_trips_u < total_trips_v:
                    # hnúturinn v fær öll þau ólíku trip_id sem u hafði
                    node_to_trip_ids[v] = node_to_trip_ids[u].copy()

                    pred[v] = u



        if found:
            leid = []
            step = target_id
            while step is not None:
                leid.insert(0, step)
                step = pred.get(step)

            return leid, final_time

        return [], float('inf')


    for source_id in source_ids:
        if str(source_id) not in g:
          continue

        for target_id in target_ids:
            if str(target_id) not in g:
              continue

            if source_id != target_id:
                path, final_time = Dijkstra3(g, str(source_id), str(target_id), departure_time_str)

                if path != [] and final_time < shortest_length:
                    shortest_path = path
                    shortest_length = final_time

    return shortest_path, shortest_length


source_stop_name = 'Meistaravellir'
target_stop_name = 'FB'
initial_departure_time = "11:30:00"


path, result_time = find_shortest_path_labb_faerri_skiptingar(g_labb.graph, source_stop_name, target_stop_name, initial_departure_time)


if result_time != float('inf'):
    prenta_leid_labb(path, g_labb.graph, result_time)
else:
    print("Engin leið fannst")
```

### 2.3.8 A*

Fyrir A* er nauðsynilegt að finna fjarlægðarfall sem er löglegt. Það sem við þurfum að hafa í huga þegar við veljum fallið er að passa að fallið gefi niðurstöðu sem er alltaf betri en raunverulega niðurstaðan frá stoppustöð yfir í lokastöð og að fallið sé minnkandi. Við getum því notað fall sem myndi skila niðurstöðu af beinni línu frá stoppustöð í aðra og gera ráð fyrir að við séum að keyra þá leið á 90 km/klst (hámarkshraði til þess að uppfylla skilyrðið c(e) + f(v) ≥ f(u) (þar sem e er leggur frá u til v og f er fjarlægðarfallið) vegna þess hvorki strætó né gangandi vegfarandi fer hraðar en 90 km/klst að meðaltali og því er skilyrðið uppfyllt)


```python
import numpy as np

def precalculate_heuristics_vectorized(stops, average_speed=90):
    # Create an array of latitudes and longitudes
    stop_ids = stops['stop_id'].unique()
    coords = stops.set_index('stop_id').loc[stop_ids][['stop_lat', 'stop_lon']].to_numpy()
    lat = np.radians(coords[:, 0])
    lon = np.radians(coords[:, 1])

    # Reikna fjarlægð með vigrum
    sin_lat = np.sin(lat[:, np.newaxis] - lat)
    sin_lon = np.sin(lon[:, np.newaxis] - lon)
    cos_lats = np.cos(lat[:, np.newaxis]) * np.cos(lat)

    a = sin_lat**2 + cos_lats * sin_lon**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance_matrix = 6371 * c

    time_matrix = (distance_matrix / average_speed) * 60

    np.fill_diagonal(time_matrix, 0)

    heuristics = {orig: {dest: time_matrix[i, j] for j, dest in enumerate(stop_ids)}
                  for i, orig in enumerate(stop_ids)}
    return heuristics

heuristic_values = precalculate_heuristics_vectorized(stops)
```

```python
def find_shortest_path_labb_A_star(g, source, target, heuristics, departure_time_str):
  shortest_path = []
  shortest_length = float('inf')

  source_ids_raw = stop_name_to_stop_ids[source]
  target_ids_raw = stop_name_to_stop_ids[target]

  # sum stop_id eru ekki í netinu
  source_ids = [stop_id for stop_id in source_ids_raw if str(stop_id) in g.graph]
  target_ids = [stop_id for stop_id in target_ids_raw if str(stop_id) in g.graph]


  def A_star_search(graph, start, end, heuristics, start_time_str):

      start_time = time_to_minutes(start_time_str)
      distances = {node: float('infinity') for node in graph.graph.keys()}
      distances[start] = start_time

      estimated_cost_dict = {}
      estimated_cost_dict[start] = start_time


      paths = {node: [] for node in graph.graph.keys()}
      paths[start] = [start]

      queue = [(start_time, start, [start], start_time)]

      while queue:
          current_cost, current_node, path, current_time = heapq.heappop(queue)

          if current_cost != estimated_cost_dict[current_node]:
              continue

          if current_node == end:
              final_time = current_time
              return path, final_time

          for neighbor, (time, mode) in graph.graph[current_node].items():
              if mode == 'bus':
                  next_time = time_to_minutes(time)
                  if next_time < current_time:
                      continue
              else:
                  next_time = current_time + time

              if next_time < distances[neighbor]:
                  distances[neighbor] = next_time
                  new_path = path + [neighbor]
                  paths[neighbor] = new_path
                  heuristic_time = heuristics.get(int(neighbor), {}).get(int(end), 0)
                  estimated_cost = next_time + heuristic_time
                  estimated_cost_dict[neighbor] = estimated_cost
                  heapq.heappush(queue, (estimated_cost, neighbor, new_path, next_time))

      return [], float('infinity')


  for source_id in source_ids:
      for target_id in target_ids:
          if source_id != target_id:
              path, final_time = A_star_search(g, str(source_id), str(target_id), heuristics, departure_time_str)

              if path != [] and final_time < shortest_length:
                  shortest_path = path
                  shortest_length = final_time

  return shortest_path, shortest_length



source_stop_name = 'Meistaravellir'
target_stop_name = 'FB'
initial_departure_time = "11:30:00"

path, final_time = find_shortest_path_labb_A_star(g_labb, source_stop_name, target_stop_name, heuristic_values, initial_departure_time)
final_time_str = minutes_to_time(final_time)

prenta_leid_labb(path, g_labb.graph, final_time)
```
